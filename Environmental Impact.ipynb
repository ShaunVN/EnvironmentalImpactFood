# Adding values

```python
#Add a % value of total emmissions
df['Total emissions %'] = df['Total_emissions']/df['Total_emissions'].sum()*100
```

```python
df.head()
```

```python
#Test to see if % column is complete, i.e. adds to 100
df['Total emissions %'].sum()
```

```python
#Re-order the new column
#Method of re-order taken from datagy.io/reorder-pandas-columns/

percentage = df['Total emissions %']
df = df.drop(columns='Total emissions %')
df.insert(loc=9,column = 'Total emissions %',value = percentage)
df.head()
```

```python
#Sort the values by total emmissions, highest to lowest
df.sort_values(by=['Total_emissions'],ascending=False,inplace=True)
df.head()
```
